```python
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, 0:4], 
                                                    df["target"], 
                                                    test_size=0.3, 
                                                    random_state=47)
```

```python
## y(class, target) value ==> categorical data ==> classfication model
## knn이라는 이름의 모델
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
print(knn)
```

```python
knn.fit(x_train, y_train)
```

```python
# 예측
y_pred = knn.predict(x_test)
y_pred
```

```python
#1. load data
#2. preprocessing
#3. train_test_split ==> hold-out dataset
#4. baseline model(a) ==> model tunning(hyper param.) : A'
#5. target model(b) ==>  model tunning(hyper param.) : B'
#6. PERFORMANCE TEST : USING METRIC(MSE, MAE(*), ACC,PERCISION, RECALL.AUC_SCORE(*))M

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_test, y_test)
acc_result = knn.score(x_test, y_test)
print(acc_result)
```

```python
knn_acc = list()

for k in range(2,20):
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(x_train, y_train)
    acc_result = knn.score(x_test, y_test)
    item = {
        "k" : k,
        "acc" : acc_result
    }
    knn_acc.append(item)
```

```python
for item in knn_acc:
    print(item)
```

```python
import matplotlib.pyplot as plt

x, y = list(), list()
for item in knn_acc:
    x.append(item["k"])
    y.append(item["acc"])

plt.plot(x,y)
plt.show()
```

```python
knn_acc = list()

acc_train = []
acc_test = []
k_list = range(1, 100, 2)
for k in k_list:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(x_train, y_train)
    knn.score(x_train, y_train)
    
    acc_train.append(knn.score(x_train, y_train))
    acc_test.append(knn.score(x_test, y_test))
```

```python
plt.plot(k_list, acc_train, 'b--', label= "acc_train")
plt.plot(k_list, acc_test, 'g', label="acc_test")
plt.ylim(0.8, 1.0)
plt.ylabel("k")
plt.ylabel("acc")
plt.legend(acc_train, acc_test)
plt.show()
```
